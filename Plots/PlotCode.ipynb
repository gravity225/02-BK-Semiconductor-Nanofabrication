## Resistance at RT

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Pfad zur Excel-Datei
C_excel_file_path = 'C:/Users/samse/Desktop/Blockkurs/02-SemiconductorNanofabrication/Results/Data/3_RT_C.xlsx'
D_excel_file_path = 'C:/Users/samse/Desktop/Blockkurs/02-SemiconductorNanofabrication/Results/Data/4_RT_D.xlsx'
B_excel_file_path = 'C:/Users/samse/Desktop/Blockkurs/02-SemiconductorNanofabrication/Results/Data/2_RT_B.xlsx'

# Daten aus der Excel-Datei lesen, wobei die erste Zeile (Header) übersprungen wird
dataC = pd.read_excel(C_excel_file_path, header=0)
dataD = pd.read_excel(D_excel_file_path, header=0)
dataB = pd.read_excel(B_excel_file_path, header=0)

# Spalte 2 (Index 1) als X-Achse und Spalte 3 (Index 2) als Y-Achse verwenden
xC = dataC.iloc[1:, 4]  # zweite Spalte (X-Achse) - erste Zeile wird übersprungen
yC = dataC.iloc[1:, 5]  # dritte Spalte (Y-Achse) - erste Zeile wird übersprungen

xD = dataD.iloc[1:, 4]  # zweite Spalte (X-Achse) - erste Zeile wird übersprungen
yD = dataD.iloc[1:, 5]  # dritte Spalte (Y-Achse) - erste Zeile wird übersprungen

xB = dataB.iloc[1:, 4]  # zweite Spalte (X-Achse) - erste Zeile wird übersprungen
yB = dataB.iloc[1:, 5]  # dritte Spalte (Y-Achse) - erste Zeile wird übersprungen


slopeC, interceptC, r_valueC, p_valueC, std_errC = linregress(xC, yC)
final_slope_C = round(slopeC * 1000, 1)
print("Steigung (slope):", final_slope_C)

slopeD, interceptD, r_valueD, p_valueD, std_errD = linregress(xD, yD)
final_slope_D = round(slopeD * 1000, 1)
print("Steigung (slope):", final_slope_D)

slopeB, interceptB, r_valueB, p_valueB, std_errB = linregress(xB, yB)
final_slope_B = round(slopeB * 1000, 1)
print("Steigung (slope):", final_slope_B)

print(yC[1])
print(yD[1])
print(yB[1])

# Plot erstellen
plt.figure(figsize=(7, 5))
plt.plot(xC, yC, linestyle='-', color='#6e8b3d', label = f'Sample C, Slope: {final_slope_C} Ohm')
plt.plot(xD, yD, linestyle='-', color='green', label = f'Sample D, Slope: {final_slope_D} Ohm')
plt.plot(xB, yB, linestyle='-', color='#A2CD5A', label = f'Sample B, Slope: {final_slope_B} Ohm')
plt.title('IV curve at RT')
plt.xlabel('Current I [muA]')
plt.ylabel('Voltage V [mV]')


plt.legend()

output_file_path = 'plot_IVcurve_RT.png'
plt.savefig(output_file_path, format='png')

plt.show()
```
